```python
# Load the dataset
df = pd.read_csv('Tweets.csv')
print(df.head())
```

```python
df.columns
```

```python
clf1 = setup(df, target='airline_sentiment', session_id=123, use_gpu=True)
```

```python
# Train a model
model = create_model('lr')
```

```python
# Finalize and save model
final_model = finalize_model(model)
```

```python
# Define a prediction function
def predict_sentiment(text):
    # Prepare data for prediction
    data = pd.DataFrame({
        'tweet_id': [None],
        'text': [text],
        'airline_sentiment_confidence': [None],
        'negativereason': [None],
        'negativereason_confidence': [None],
        'airline': [None],
        'airline_sentiment_gold': [None],
        'name': [None],
        'negativereason_gold': [None],
        'retweet_count': [None],
        'tweet_coord': [None],
        'tweet_created': [None],
        'tweet_location': [None],
        'user_timezone': [None]
    })

    print("Input data for prediction:", data)  # Debugging line
    prediction = predict_model(final_model, data=data)
    print("Prediction output:", prediction)  # Debugging line
    sentiment = prediction['prediction_label'].values[0]  # Corrected this line

    return f"The predicted sentiment is: {sentiment}"
```

```python
# Test prediction without Gradio
test_input = "Today is the best day!"
data = pd.DataFrame({
    'tweet_id': [None],
    'text': [test_input],
    'airline_sentiment_confidence': [None],
    'negativereason': [None],
    'negativereason_confidence': [None],
    'airline': [None],
    'airline_sentiment_gold': [None],
    'name': [None],
    'negativereason_gold': [None],
    'retweet_count': [None],
    'tweet_coord': [None],
    'tweet_created': [None],
    'tweet_location': [None],
    'user_timezone': [None]
})

prediction = predict_model(final_model, data=data)
print(prediction)  # Check output here
```

```python
# Set up Gradio interface
iface = gr.Interface(
    fn=predict_sentiment,
    inputs=gr.Textbox(lines=2, placeholder="Enter a tweet..."),
    outputs="text",
    title="Airline Sentiment Analysis",
    description="Enter a tweet to analyze its sentiment regarding an airline."
)
```

```python
# Launch the interface
iface.launch()
```
