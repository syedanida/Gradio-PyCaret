```python
import gradio as gr
from transformers import pipeline
```

```python
# Load a pre-trained image classification model from Hugging Face
classifier = pipeline("image-classification")
```

```python
# Define a function that will use the model to classify the input image
def classify_image(image):
    # Use the Hugging Face model to classify the image
    result = classifier(image)
    label = result[0]['label']  # Get the label of the top result
    confidence = result[0]['score']  # Get the confidence score of the top result
    return f"Prediction: {label}, Confidence: {confidence:.2f}"
```

```python
# Set up the Gradio interface
iface = gr.Interface(
    fn=classify_image,
    inputs=gr.Image(type="pil"),  # Accepts image input
    outputs="text",  # Displays text output
    title="Image Classification",
    description="Upload an image and get the predicted class!"
)
```

```python
# Launch the Gradio app
iface.launch()
```
